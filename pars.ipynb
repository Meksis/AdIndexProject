```python
import requests

headers = {
    'Authorization': 'OAuth y0_AgAAAABxT5u9AArjtwAAAADzHTIKv-VGe5uzTHqKqd3WWTTeIiZQ2WU',
}

params = {
    'id': '22386646',
    'metrics': 'ym:s:pageviews',  # Метрика для просмотров страниц
    'dimensions': 'ym:s:referer',  # Размерность для получения ссылок
}

response = requests.get('https://api-metrika.yandex.net/stat/v1/data', params=params, headers=headers)

# Проверяем успешность запроса
if response.status_code == 200:
    data = response.json()['data']
    for item in data:
        # Выводим ссылку и количество просмотров
        url = item['dimensions'][0]['name']
        pageviews = item['metrics'][0]
        print(f'Ссылка: {url}, Просмотры: {pageviews}')
else:
    print(f'Ошибка {response.status_code}: {response.text}')
```
