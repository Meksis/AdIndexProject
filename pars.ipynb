```python
way = {}
for visitors in raw_df['data'] :
    sites_metric = {}
    visitors['dimensions']
    sites_metric['Источник']= visitors['dimensions'][0]['favicon']
    sites_metric['Посетители'] = visitors['metrics'][0]
    sites_metric['читатели'] = visitors['metrics'][1]
    sites_metric['уникальные пользователи'] = visitors['metrics'][2]
    sites_metric['недочитано'] = visitors['metrics'][3]
    sites_metric['глубина просмотра'] = visitors['metrics'][4]
    sites_metric['Ср время на сайти (сек)'] = visitors['metrics'][0]


    way['Метрики порихода по доменам']=[]
    way['Метрики порихода по доменам'].append(sites_metric)

totalmetric={
'Посетители': raw_df['totals'][0],
  'читатели': raw_df['totals'][1],
  'уникальные пользователи': raw_df['totals'][2],
  'недочитано': raw_df['totals'][3],
  'глубина просмотра': raw_df['totals'][4],
  'Ср время на сайти (сек)': raw_df['totals'][5]
}

way['общие метрики']=totalmetric
```
